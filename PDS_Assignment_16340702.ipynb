A) Look for the missing values in all the columns and either impute them (replace with mean,
median, or mode) or drop them. Justify your action for this task.

B)Take off the units from some of the qualities and leave only the numerical numbers (remove CC from "Engine," lakh from "New Price," bhp from "Power," and kmpl from "Mileage").

```python
#importing all the required libraries
import pandas as pd
import numpy as np
from sklearn.impute import SimpleImputer
from datetime import datetime
```

```python
#Reading the dataset 
data = pd.read_csv("train.csv")
```

```python
data.head()
```

```python
#checking for null values in the data frame
null_data = data.isnull().sum()
print(null_data)
```

```python
#dropping New_Price column because it has more NULL values compared to other numeric columns
data = data.drop('New_Price', axis=1)
```

```python
data.head()
```

```python
#removing characters after interger values in the above columns
data['Mileage']=data['Mileage'].str.split(' ').str[0]
data['Engine']=data['Engine'].str.split(' ').str[0]
data['Power']=data['Power'].str.split(' ').str[0]
data['Seats']=data['Seats'].fillna(0)
```

```python
# converting string columns values to float 
data['Mileage'] = data['Mileage']. astype(float)
data['Engine'] = data['Engine']. astype(float)
data['Power'] = data['Power']. astype(float)
data['Seats'] = data['Seats'] . astype(float)
```

```python
#replacing the NULL values with mean

numeric_columns = ['Mileage', 'Engine', 'Power', 'Price','Seats']
imputer = SimpleImputer(strategy='mean')
data[numeric_columns] = imputer.fit_transform(data[numeric_columns])
```

```python
#checking for null values in the data frame after dropping and adding mean values in the place of NULL 
data.isnull().sum()
```
